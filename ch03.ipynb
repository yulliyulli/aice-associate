## SECTION 01. 데이터 프레임 확인하기

데이터 프레임 내 처음 n줄 데이터 `head(n)`

```python
import pandas as pd
flight = pd.read_csv('./dataset/Clean_Dataset.csv')
```

```python
flight.head() #default n=5
```

데이터 프레임 내 마지막 n줄 데이터 `tail(n)`

```python
flight.tail() #default n=5
```

데이터 프레임 속성값 `shape`
- 행과 열의 개수를 튜플 형태로 변환해서 줌

```python
flight.shape
```

데이터 프레임을 구성하는 칼럼명을 확인 `colums`

```python
flight.columns
```

데이터 유형, 각 컬럼의 데이터 수 `info()`
- 데이터 구성이나 결측치 유무를 확인

```python
flight.info()
```

데이터 컬럼별 요약 통계량 `describe()`
- count: 데이터 개수, mean: 평균값, std: 표준편차, min: 최솟값, max: 최댓값, 4분위값
- float, int 숫자형 데이터의 통계치를 확인
- 범주형 데이터 : 데이터 개수, 최빈값, 최빈값 개수를 제공

```python
flight.describe()
```

```python
flight.describe(include='all') # 범주형 데이터 포함
```

데이터프레임이 가지는 데이터 형태의 종류 `dtypes`

```python
flight.dtypes
```

지정된 컬럼의 값의 발생 횟수 `value_counts()`

```python
flight.value_counts()
```
