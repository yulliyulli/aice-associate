## SECTION 03. 데이터프레임 변형하기

그룹화 하기 `groupby`
- 제공 정보
    - count: 데이터 개수, size: 집단별 크기, sum: 데이터의 합, mean,std,var: 평균,표준편차,분산, min-max: 최솟값-최댓값

```python
import pandas as pd
flight = pd.read_csv('./dataset/Clean_Dataset.csv')
```

```python
airline_group = flight.groupby('airline')
airline_group # 데이터 프레임으로 그룹까지 생성
```

```python
airline_group.groups # 그룹화된 결과 확인
```

```python
airline_group.count() # 데이터 개수 확인
```

인덱스로 그룹화하기
- `set_index` 컬럼을 인덱스로 변경, 기존의 인덱스를 제거하고 컬럼 중 하나를 인덱스로 설정
- `reset_index` 인덱스 초기화

```python
flight.set_index(['airline', 'arrival_time']) # index 를 설정한 결과, 그룹핑은 안된 상태
```

```python
# level(depth)으로 0으로 설정한 인덱스 중 airline을 기준으로 그룹화
flight.set_index(['airline', 'arrival_time']).groupby(level=[0]).min() 
```

```python
# level(depth)으로 airline, arrival_time을 기준으로 그룹화
flight.set_index(['airline', 'arrival_time']).groupby(level=[0,1]).min() 
```

aggregate로 집계하기
- 값을 다양하게 집계하여 한번에 볼 수 있음

```python
import numpy as np
flight.set_index(['airline', 'arrival_time']).groupby(level=[0,1]).aggregate([np.min, np.max])
```

피벗테이블 생성하기
- `pivot` 값이 중복되는 경우 에러가 발생
    - ValueError: Index contains duplicate entries, cannot reshape
- `pivot_table` 값이 중복되는 경우 처리를 지정할 수 있는 `aggfunc` 매개변수를 제공

```python
# 데이터프레임 생성
df = pd.DataFrame({
    'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 20, 30, 40]
})

df.pivot(index='Date', columns='Category', values='Value')
```

```python
# 피벗 테이블 생성
pivot_table = df.pivot_table(index='Date', columns='Category', values='Value', aggfunc='sum')
print(pivot_table)
```

## SECTION 04. 데이터프레임 병합하기

2개 이상의 데이터 프레임을 병합 `concat`
- 칼럼 명이 같은 경우
    - `ignore_index` 기존 인덱스를 유지하거나 무시
        - `False(default)` 기존 인덱스 유지
        - `True` 기존 인덱스 무시(인덱스 재배열)
    - `axis` 데이터프레임을 행 레벨로 병합할 것인지, 칼럼 레벨로 병합할 것인지 정의
        - `axis=0(default)` 행 레벨로 병합(위+아래, 세로로 합치기)
        - `axis=1` 칼럼 레벨로 병합(왼쪽+오른쪽, 가로로 합치기)
- 칼럼 명이 다른 경우
    - `join`
        - `outer` 합집합
        - `inner` 교집합
- 인덱스가 중복인 경우
    - `verify_integrity` 데이터프레임에서 인덱스의 중복 여부를 점검하는 파라미터
        - `False(default)` 인덱스 중복으로 인한 에러가 발생하지 않음
        - `True` 인덱스 중복으로 인한 에러 발생

특정한 key 기준으로 데이터프레임 병합 `merge/join`
- `merge` 서로 다른 형태와 값의 데이터프레임을 병합할 때 사용
    - `on` 어떤 칼럼을 이용하여 merge 할 것인지 정의
    - `how` merge 방식을 정의
        - `inner(default)` 일치하는 값이 있는 경우에만 데이터를 가져옴.
        - `left` 왼쪽 데이터프레임을 기준으로 오른쪽 데이터프레임을 병합
        - `right` 오른쪽 데이터프레임을 기준으로 왼쪽 데이터프레임을 병합
        - `outer` left 와 right를 합한 데이터프레임을 병합
